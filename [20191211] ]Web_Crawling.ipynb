### OPEN API
- api로 전국 골프장 현황 불러오기

```python
###open Api 불러오기#####
serviceKey ="2ae0dbe8949e48e192f2f83b837360df"
endpoint = 'https://openapi.gg.go.kr/GolfPlace?KEY={}'.format(serviceKey)

resp=requests.get(endpoint)
html=resp.text

####html parsing#####
soup=BeautifulSoup(html, 'html.parser')

golfplace_name=soup.find_all('bizplc_nm')
golf_address=soup.find_all('refine_roadnm_addr')

####주소 DataFrame 만들기####
golfplace_name_list=[]
golf_address_list=[]

for i in np.arange(0,len(golf_address)):
    golfplace_name_list.append(golfplace_name[i].text)
    golf_address_list.append(golf_address[i].text)
    
df=pd.DataFrame()
df['Golf_Place']=golfplace_name_list
df['Golf_Address']=golf_address_list
```
